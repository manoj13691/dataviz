```python
import pandas as pd
from collections import Counter
import json
import math
import scipy
import operator
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
from collections import Counter
import numpy as np
import wordcloud
# Hide all the warnings
import warnings
warnings.filterwarnings('ignore')

canada_data = pd.read_csv("./data/CAvideos.csv")
germany_data = pd.read_csv("./data/DEvideos.csv")
france_data = pd.read_csv("./data/FRvideos.csv")
great_britain_data = pd.read_csv("./data/GBvideos.csv")
us_data = pd.read_csv("./data/USvideos.csv")
```

```python
files = ['./data/CAvideos.csv', './data/DEvideos.csv', './data/FRvideos.csv', './data/GBvideos.csv', './data/USvideos.csv']
df_fin = list()
for i in files:
    df_init = pd.read_csv(i)
    df_init['country'] = i[:2]
    df_fin.append(df_init)

vids = pd.concat(df_fin)
```

```python
# For Canada data
canada_trending_categories = dict()
canada_category_ids = list(canada_data["category_id"])

for i in canada_category_ids:
    if i not in canada_trending_categories:
        canada_trending_categories[i] = 1
    else:
        canada_trending_categories[i] +=1

sorted_dictionary = Counter(canada_trending_categories)
#print(sorted_dictionary)

# Top 5 category ID's 
top_5_categories = sorted_dictionary.most_common()
#print(top_5_categories)

with open('./data/CA_category_id.json', 'r') as f:
    canada_category_json = json.load(f)


cat = canada_category_json["items"]
id_category_mapping = {}
for i in range(len(cat)):
    id_ = cat[i]["id"]
    title = cat[i]["snippet"]["title"]
    if id_ not in id_category_mapping:
        id_category_mapping[id_] = title    

# Trending categories in Canada
category_count_visualization_dict = {}
ca_category_dict = {}
for i in top_5_categories:
    id_ = str(i[0])
    count = str(i[1])
    if id_ in id_category_mapping:
        category_count_visualization_dict[id_category_mapping[id_]] = int(count)
        ca_category_dict[id_category_mapping[id_]] = int(count)
        print(id_category_mapping[id_]+", "+"Count:"+count)
    else:
        print("Unknown, "+"Count:"+count)


# Sort the dictionary based on value
category_count_visualization_wc = category_count_visualization_dict
category_count_visualization_dict = sorted(category_count_visualization_dict.items(), key = operator.itemgetter(1))
```

```python
# Word Cloud
wrd_cnt = Counter(category_count_visualization_wc)
wc = wordcloud.WordCloud(width=1000, height=500)
wc.generate_from_frequencies(wrd_cnt)
plt.figure(figsize=(20,10))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```

```python
# Bar Graph
x = []
y = []
for category, count in category_count_visualization_dict:
    x.append(category)
    y.append(count)


plt.figure(figsize=(10,10))
plt.barh(x,np.log10(y))
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.barh(x,y)
plt.show()
```

```python
all_categories = set()

for category in ca_category_dict:
    all_categories.add(category)

for category in ge_category_dict:
    all_categories.add(category)
    
for category in fr_category_dict:
    all_categories.add(category)

for category in gb_category_dict:
    all_categories.add(category)

for category in us_category_dict:
    all_categories.add(category)

all_categories = list(all_categories)
```

```python
all_country_vectors = []
all_country_vectors.append(all_categories)

def get_country_vectors(country_category_dict):
    country_vector = []
    for category in all_categories:
        if category in country_category_dict:
            country_vector.append(country_category_dict[category])
        else:
            country_vector.append(0)
    
    #country_vector = preprocessing.normalize([country_vector])
    #return list(country_vector[0])
    return list(country_vector)
        
all_country_vectors.append(get_country_vectors(ca_category_dict))
all_country_vectors.append(get_country_vectors(ge_category_dict))
all_country_vectors.append(get_country_vectors(fr_category_dict))
all_country_vectors.append(get_country_vectors(gb_category_dict))
all_country_vectors.append(get_country_vectors(us_category_dict))
```

```python
col_list = all_country_vectors.pop(0)
col_list
all_country_vectors = np.array(all_country_vectors)
```

```python
col_list
```

```python
all_country_vectors
```

```python
cosine_similarities = []
for i in range(len(all_country_vectors)):
    similarity_vector = []
    for j in range(len(all_country_vectors)):
        similarity = 1-scipy.spatial.distance.cosine(all_country_vectors[i], all_country_vectors[j])
        similarity_vector.append(similarity)
        
    cosine_similarities.append(similarity_vector)
    
cosine_similarities = np.array(cosine_similarities)
```

```python
cosine_similarities
```

```python
X = cosine_similarities
fig, ax = plt.subplots(figsize = (10, 10))
hmap = ax.imshow(X, cmap = 'YlOrBr', interpolation = 'nearest')
labs = ['']
countries = ['CA', 'DE', 'FR', 'GB', 'US']
cbar = fig.colorbar(hmap)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        text = ax.text(j, i, round(X[i, j], 2), ha = "center", va = "center", color = "black")
for i in countries:
    labs.append(i)
ax.set_yticklabels(labs)
ax.set_xticklabels(labs)
```

```python
category_from_json={}
with open("./data/US_category_id.json","r") as file:
    data=json.load(file)
    for category in data["items"]:
        category_from_json[category["id"]]=category["snippet"]["title"]#it Stores the category id with category name
```

```python
#Lets change the data-format of following fields to int type, which will be easy to use.
df_USA = us_data
list1=["views likes dislikes comment_count".split()] 
for column in list1:
    df_USA[column]=df_USA[column].astype(int)
#Similarly Convert The Category_id into String,because later we're going to map it with data extracted from json file    
list2=["category_id"] 
for column in list2:
    df_USA[column]=df_USA[column].astype(str)
df_USA["Category"]=df_USA["category_id"].map(category_from_json) #category_from_json{} stores categories from JSON files
```

```python
df_USA.shape
```

```python
df_USA = df_USA.loc[df_USA['Category'].isin(['Entertainment','Music', 'Howto & Style', 'Comedy', 'People & Blogs'])]
```

```python
months = []
trending_date = df_USA["trending_date"]
for date in trending_date:
    month = date.split(".")[2]
    months.append(month)
print(len(months))
```

```python
df_USA["Month"] = months
df_USA.head(5)
```

```python
month_category_count = {}
category = list(df_USA["Category"])
month = list(df_USA["Month"])

for i in range(len(month)):
    key = (month[i], category[i])
    if key not in month_category_count:
        month_category_count[key] = 1
    else:
        month_category_count[key] +=1
```

```python
months = []
categories = []
counts = []

for i in month_category_count:
    months.append(i[0])
    categories.append(i[1])
    counts.append(month_category_count[i])
```
